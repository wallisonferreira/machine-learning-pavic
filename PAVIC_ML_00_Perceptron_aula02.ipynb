```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import random
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
#Algoritimo perceptron

# Inicializar o 'w' e o 'b'
# Para cada amostra
  # calcular o y_pred = f(E xi*wi + b)onde f é  função de ativação step, para regressão trocamos para linear
  # calcular o erro erro = y - y_pred
  # atualizar os pesos wi = wi + lambda * erro * xi
  # atualizar o bias bi = bi + lambda*erro
# repetir até encontrar o menor erro
```

# Classificação

## Porta OR ou AND

```python
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
#y = np.array([0, 1, 1, 1]) # porta OR
y = np.array([0, 0, 0, 1]).T # porta AND

print(x.shape, y.shape)
```

```python
#Initilize w and b
D = x.shape[1] # qtd entradas
w = 2*np.random.random(D)-1
b = 2*np.random.random()-1
w,b
```

```python
#LOOP TRAIN - PUT YOU CODE HERE
for step in range(100):
    learning_rate = 1.0
    cost = 0
    for x_n, y_n in zip(x, y):
        y_pred = np.dot(x_n, w) + b
        y_pred = np.where(y_pred > 0, 1, 0) # activation function # where is > 0 turns 1 else 0
        error = y_n - y_pred # calculates error
        w = w + learning_rate * np.dot(error, x_n) # updates weights of objective function / lambda is the learning rate
        b = b + learning_rate * error
        cost += error ** 2
    if step % 10 == 0:
        print('Step {0}: {1}'.format(step, cost))

    print('w', w)
    print('b', b)
    print( 'y_pred: {0}'.format( np.dot( x, np.array(w) ) + b ) )
```

```python
def plot_linear_classifier(x, y, w, b):
    x1_min, x1_max = x[:,0].min(), x[:,0].max()
    x2_min, x2_max = x[:,1].min(), x[:,1].max()

    x1, x2 = np.meshgrid(np.linspace(x1_min-1, x1_max+1,100), np.linspace(x2_min-1, x2_max+1, 100))
    x_mesh = np.array([x1.ravel(), x2.ravel()]).T

    plt.scatter(x[:,0], x[:,1], c=y.ravel(), cmap='bwr')

    y_mesh = np.dot(x_mesh, np.array(w).reshape(1, -1).T) + b
    y_mesh = np.where(y_mesh <= 0, 0, 1)

    plt.contourf(x1, x2, y_mesh.reshape(x1.shape), cmap='bwr', alpha=0.5)
    plt.xlim(x1_min-1, x1_max+1)
    plt.ylim(x2_min-1, x2_max+1)
```

```python
#DEFINE W AND B - PUT YOU CODE HERE
D = x.shape[1]
w = 2*np.random.random(D)-1
b = 2*np.random.random()-1

w, b
```

```python
#LOOP TRAIN - PUT YOU CODE HERE
for step in range(100):
    learning_rate = 1.0
    cost = 0
    for x_n, y_n in zip(x, y):
        y_pred = np.dot(x_n, w) + b
        y_pred = np.where(y_pred > 1, 1, 0)
        error = y_n - y_pred
        w = w + learning_rate * np.dot(error, x_n)
        b = b + learning_rate * error
        cost += error ** 2
    if step % 10 == 0:
        print('Step {0}: {1}'.format(step, cost))

print('w', w)
print('b', b)
print( 'y_pred: {0}'.format( np.dot( x, np.array(w) ) + b ) )
```

```python
plot_linear_classifier(x, y, w, b)
```

# Perceptron Regressão linear

Para transformar o Perceptron em um __regressor linear__, só o que temos de fazer é __remover a função de ativação _step___, transformando-a em uma função de ativação linear.

Apesar dessa modificação, __a fórmula de atualização dos pesos não sofre nenhuma alteração__.

Vamos, então, implementar nosso perceptron para classificação em Python, Numpy, Keras e TensorFlow:

```python
df = pd.read_csv('/content/medidas.csv')
print(df.shape)
df.head(10)
```

```python
x = df.Altura.values
y = df.Peso.values

plt.figure()
plt.scatter(x, y)
plt.xlabel('Altura')
plt.ylabel('Peso')
```

```python
print(x.shape, y.shape)
```

```python
x = x.reshape(-1, 1)

print(x.shape, y.shape)
```

```python
#LOOP TRAIN - PUT YOU CODE HERE
D = x.shape[1]
w = 2 * np.random.random(D)-1
b = 2 * np.random.random()-1
lr = 1e-7
for step in range(10001):
    cost = 0
    for x_n, y_n in zip(x, y):
        y_pred = np.dot(x_n, w) + b
        #y_pred = np.where(y_pred > 1, 1, 0) # na regressão, tiramos a função de ativação
        error = y_n - y_pred
        w = w + lr * np.dot(error, x_n)
        b = b + 1e-2 * error
        cost += error ** 2
    if step % 1000 == 0:
        print('Step {0}: {1}'.format(step, cost))
    print('###################################')
    print('w', w)
    print('b', b)
    print('error: ', error)
    print('###################################')
```

```python
# os dados nao estao normalizados peso em kg e altura em m
# vamos normalizar
# PUT YOU CODE HERE
minmax = MinMaxScaler(feature_range=(-1, 1))
x = minmax.fit_transform(x.astype(np.float64))
```

```python
plt.figure()
plt.scatter(x, y)
plt.xlabel('Altura')
plt.ylabel('Peso')
```

```python
reg = LinearRegression()
reg.fit(x,y)

print('w: ', reg.coef_)
print('b: ', reg.intercept_)
```

```python
def plot_line(w, b):
    plt.figure(0, figsize=(8,8))
    plt.subplot(1,1,1)
    plt.scatter(x, y)
    y_pred = x*w + b
    plt.plot(x, y_pred, c='red')
    plt.xlim(140, 210)
    plt.ylim(40, 120)

    plt.subplot(1,1,1)
    x_ = np.array([0, x.max()])
    y_ = x_*w + b
    plt.scatter(x, y)
    plt.plot(x_, y_, c='red')
    plt.xlim(0, 1)
    plt.ylim(-160, 120)

    plt.show()
```
